# Detect Corners

```python
import numpy as np
import cv2 as cv
from matplotlib import pyplot as plt
from config import config as cfg
import os

#fpath = os.path.join(cfg.IMAGES_DIR,'blox.jpg')
fpath = os.path.join(cfg.IMAGES_DIR,'features','IMG_1262_crop.JPG')
im = cv.imread(fpath)

gray = cv.cvtColor(im,cv.COLOR_BGR2GRAY)

corners = cv.goodFeaturesToTrack(gray,20,0.01,10)
corners = np.int0(corners)

dst = im.copy()

for i in corners:
    x,y = i.ravel()
    cv.circle(dst,(x,y),45,(255,0,0,100),-1)

plt.figure(figsize=(12,8))
dst_rgb = cv.cvtColor(dst,cv.COLOR_BGR2RGB)
plt.yticks([]),plt.xticks([])
plt.imshow(dst_rgb),plt.show()
```
