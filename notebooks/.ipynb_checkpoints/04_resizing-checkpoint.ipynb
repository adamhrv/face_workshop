## Load random image using OpenCV

```python
dir_images = app_cfg.DIR_IMAGES
filepaths = glob(join(dir_images, '*.jpg'))
```

```python
rn = random.randint(0, len(filepaths))
print('rn', rn)
fp_im = filepaths[rn]
print('fp_im', fp_im)
```

```python
im_bgr = cv.imread(fp_im)
im_rgb = cv.cvtColor(im_bgr ,cv.COLOR_BGR2RGB)
plt.imshow(im_rgb)
```

## Resize image

```python
im_rgb_resized = imutils.resize(im_rgb, width=200)
plt.imshow(im_rgb_resized)
```

```python
im_rgb_resized = imutils.resize(im_rgb, width=2000)
plt.imshow(im_rgb_resized)
```

```python
plt.figure(figsize=(16,8))
plt.imshow(im_rgb_resized)
```
