# 01: Loading, Displaying, and Saving Images

This notebook demonstrates loading, displaying, and saving images

- 02a: Load image with OpenCV
- 02b: Read image properties
- 02c: Display image
- 02d: Save image



```python
import os
import os.path as osp
import cv2
%matplotlib inline
import matplotlib.pyplot as plt
```

```python
# Setup directory paths
project_root = osp.dirname(os.path.realpath(''))
DATA_STORE = osp.join(project_root,'data_bodega')

# Asset directories
IMAGES_DIR = osp.join(DATA_STORE,'images')
print(IMAGES_DIR)
# OUTPUT_DIR = osp.join(DATA_STORE,'output')
# MODELS_DIR = osp.join(DATA_STORE,'models')
```

```python
# Plot images inline using Matplotlib
def pltimg(im,title=None,mode='rgb',figsize=(8,12),dpi=160,output=None):
    plt.figure(figsize=figsize)
    plt.xticks([]),plt.yticks([])
    if title is not None:
        plt.title(title)
    if mode.lower() == 'bgr':
        im = cv2.cvtColor(im,cv2.COLOR_BGR2RGB)
    f = plt.gcf()
    plt.imshow(im)
    plt.show()
    plt.draw()
    if output is not None:
        bbox_inches='tight'
        ext=osp.splitext(output)[1].replace('.','')
        f.savefig(output,dpi=dpi,format=ext)
        print('Image saved to: {}'.format(output))
# This function could also be placed in an external file 
```

## 02a Load an image

```python
# Read an image and display it
filepath = osp.join(IMAGES_DIR,'obama.jpg')

# Read it (will be in BGR, not RGB)
im = cv2.imread(filepath)
```

## 02b Image properties

```python
width = im.shape[1] # width and height are reversed
height = im.shape[0]
print('Width: {}, Height: {}'.format(width,height))

print("Image shape: {}".format(im.shape))
```

## 02c Dislpay Image

```python
pltimg(im) # this assumes image is in RGB colorspace, but it's not
```

```python
# The imag will appear blue because R and B color channels are switched
# By default OpenCV uses BGR (because it's faster) but we need RGB
# OpenCV has a built-in color conversion function to fix this
# Convert the image from BGR to RGB

im_rgb = cv2.cvtColor(im,cv2.COLOR_BGR2RGB)
# Or set the mode to 'bgr' using the pltimg fuction
pltimg(im_rgb,title="HELLO! BGR --> RGB")
```

```python
im_r = im.copy()
im_r[:,0] = 0
im_r[:,1] = 0
pltimg(im_r, mode='bgr', title='mode=bgr')
```

```python
# Or you can set the mode in the pltimg fuction
pltimg(im, mode='bgr', title='mode=bgr')
```

```python
# If you want, save a copy
fpath_out = osp.join(OUTPUT_DIR,'ch1_2.jpg')
cv2.imwrite(fpath_out,im)
```

```python
# You can also save it as a PDF or PNG
output_filepath = osp.join(OUTPUT_DIR,'ch1_1.pdf')
pltimg(im, mode='bgr',title='Test output',output=output_filepath)
```
