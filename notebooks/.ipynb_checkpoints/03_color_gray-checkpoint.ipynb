# Color vs Grayscale

Demonstrate that color can represent one image to a human perceiver while presenting a different image to the grayscale transformed image processing algorithm.

Grayscale is g = 0.299r + 0.587g + 0.114b

http://www.had2know.com/technology/rgb-to-gray-scale-converter.html


The weights used in this conversion are calibrated for contemporary CRT phosphors: Y = 0.2125 R + 0.7154 G + 0.0721 B

```python
import os
from os.path import join
%matplotlib inline
import matplotlib
import numpy as np
import matplotlib.pyplot as plt
from matplotlib import gridspec
plt.style.use('ggplot')
import numpy as np
import cv2
# from skimage import color
# from skimage import io
from PIL import Image, ImageDraw
from random import randint
import itertools
import math
# import svgwrite
import datetime
from time import time
```

```python
opts = itertools.permutations(['R','G','B'])
opts = [''.join(a) for a in opts]
```

```python
def hex_to_rgb(value):
    value = value.lstrip('#')
    lv = len(value)
    return tuple(int(value[i:i+lv/3], 16) for i in range(0, lv, lv/3))

def rgb_to_hex(rgb):
    return '#%02x%02x%02x' % rgb
```

```python
def g2c(val):
    '''
    Tends to return pinkish/blue values
    b: brightness value -0255
    returns r,g,b
    '''
    # coeffecients
    cr, cg, cb = (0.299,0.587,0.114)
    crmax, cgmax, cbmax = (cr*255,cg*255,cb*255)
    #print crmx, cgmx, cbmx
    if val < cbmax:
        result = (0,0,val//cb)
    elif val < cbmax + crmax:
        val_adj = val - cbmax
        result = (val_adj//cr,0,cbmax)
    else:
        val_adj = val - cbmax - crmax
        result = (crmax//cr,val_adj//cg, cbmax//cb)
    return tuple([int(r) for r in result])
```

```python
def g2cr(val,opt=None):
    '''
    b: brightness value -0255
    returns r,g,b
    '''
    # coeffecients
    cr, cg, cb = (0.299,0.587,0.114)
    crmax = math.floor(cr*255)
    cgmax = math.floor(cg*255)
    cbmax = math.floor(cb*255)
    
    if opt is None:
        opt = opts[randint(0,len(opts))-1]
        
    if opt =='GRB':
        # Green
        gvmax = val - int(crmax + cbmax)
        ming = int(min(255,max(0,gvmax//cg)))
        maxg = int(min(255,val//cg))
        rg = int(min(255,randint(ming,maxg)))
        val_adj = val - int(rg*cg)

        # Red
        rvmax = val_adj - int(cbmax)
        minr = min(255,rvmax//cr)
        maxr = min(255,val_adj//cr)
        rr = randint(minr,maxr)
        val_adj = val_adj - int(rr*cr)

        # Blue
        rb = int(min(255,val_adj//cb))
            
    elif opt == 'GBR':
        # Green
        gvmax = val - int(crmax + cbmax)
        ming = int(min(255,max(0,gvmax//cg)))
        maxg = int(min(255,val//cg))
        rg = int(min(255,randint(ming,maxg)))
        val_adj = val - int(rg*cg)
    
        # blue
        bvmax = val_adj - int(crmax)
        minb = min(255,bvmax//cb)
        maxb = min(255,val_adj//cb)
        rb = randint(minb,maxb)
        val_adj -= int(rb*cb)
        
        # Red
        rr = int(min(255,val_adj//cr))

    elif opt == 'RGB':
        # red
        rvmax = val - int(cgmax + cbmax)
        minr = int(min(255,max(0,rvmax//cr)))
        maxr = int(min(255,val//cr))
        rr = int(min(255,randint(minr,maxr)))
        val_adj = val - int(rr*cr)

        # green
        gvmax = val_adj - int(cbmax)
        ming = min(255,gvmax//cg)
        maxg = min(255,val_adj//cg)
        rg = randint(ming,maxg)
        val_adj = val_adj - int(rg*cg)

        # Blue
        rb = int(min(255,val_adj//cb))

    elif opt == 'RBG':
        # red
        rvmax = val - int(cgmax + cbmax)
        minr = int(min(255,max(0,rvmax//cr)))
        maxr = int(min(255,val//cr))
        rr = int(min(255,randint(minr,maxr)))
        val_adj = val - int(rr*cr)

        # blue
        bvmax = val_adj - int(cgmax)
        minb = min(255,bvmax//cb)
        maxb = min(255,val_adj//cb)
        rb = randint(minb,maxb)
        val_adj = val_adj - int(rb*cb)

        # green
        rg = int(min(255,val_adj//cg))

    elif opt == 'BGR':
        # Blue
        bvmax = val - int(cgmax + crmax)
        minb = int(min(255,max(0,bvmax//cb)))
        maxb = int(min(255,val//cb))
        rb = int(min(255,randint(minb,maxb)))
        val_adj = val - int(rb*cb)

        # Green
        gvmax = val_adj - int(crmax)
        ming = min(255,gvmax//cg)
        maxg = min(255,val_adj//cg)
        rg = int(min(255,randint(ming,maxg)))
        val_adj = val_adj - int(rg*cg)

        # Red
        rr = int(min(255,val_adj//cr))

    elif opt == 'BRG':
        # Blue
        bvmax = val - int(cgmax + crmax)
        minb = int(min(255,max(0,bvmax//cb)))
        maxb = int(min(255,val//cb))
        rb = int(min(255,randint(minb,maxb)))
        val_adj = val - int(rb*cb)

        # Red
        rvmax = val_adj - int(cgmax)
        minr = min(255,rvmax//cr)
        maxr = min(255,val_adj//cr)
        rr = int(min(255,randint(minr,maxr)))
        val_adj = val_adj - int(rr*cr)

        # Green
        rg = int(min(255,val_adj//cg))
        
    # FIXME
    rr = max(0,rr)
    rg = max(0,rg)
    rb = max(0,rb)
    return tuple([int(rr), int(rg), int(rb)])
```

```python
# Validate fun
for i in range(100):
    ri = randint(0,255)
    g2cr(ri)
```

```python
# Create a rectangle with four colors
im = Image.new('RGB',(200,200),rgb_to_hex((0,0,0)))
dr = ImageDraw.Draw(im)
dr.rectangle(((0,0),(100,100)), fill='#00aa00')
dr.rectangle(((100,0),(200,100)), fill='#e63500')
dr.rectangle(((0,100),(100,200)), fill='#5053f0')
dr.rectangle(((100,100),(200,200)), fill='#e60ec8')
del dr # clear drawing

# convert 4-color rectangle to gray, then back to rgb
imnp = np.array(im).astype(np.uint8)
imnpg = cv2.cvtColor(imnp,cv2.COLOR_RGB2GRAY)
imnpg = cv2.cvtColor(imnpg,cv2.COLOR_GRAY2RGB)

# plot
plt.figure(figsize=(12,8))
plt.subplot(1, 2, 1) # rows, coloumns, index
plt.title('RGB')
plt.xticks([]),plt.yticks([])
plt.imshow(imnp,vmin=0,vmax=255)

plt.subplot(1, 2, 2)
plt.title('RGB --> Gray --> RGB')
plt.xticks([]),plt.yticks([])
plt.imshow(imnpg,cmap='gray',vmin=0,vmax=255)

# Save fig
#figimg = plt.gcf() 
#figimg.savefig('color_grayscale.png',dpi=300, bbox_inches='tight')

plt.show()
```

```python
# Start as RGB
im_pil_rgb = Image.new('RGB',(200,200),'#000000')
dr = ImageDraw.Draw(im_pil_rgb)

dr.rectangle(((0,0),(100,100)), fill='#00aa00')
dr.rectangle(((100,0),(200,100)), fill='#e63500')
dr.rectangle(((0,100),(100,200)), fill='#5053f0')
dr.rectangle(((100,100),(200,200)), fill='#e60ec8')
del dr

im_np_rgb = np.array(im_pil_rgb).astype(np.uint8)
im_np_rgb_gr = cv2.cvtColor(im_np_rgb,cv2.COLOR_RGB2GRAY)

# original image to YCrCb -> Gray -> BGR
im_np_ycrcb = cv2.cvtColor(im_np_rgb,cv2.COLOR_RGB2YCR_CB)
im_np_ycrcb_gray = cv2.cvtColor(im_np_ycrcb,cv2.COLOR_YCR_CB2RGB)
im_np_ycrcb_gray = cv2.cvtColor(im_np_ycrcb_gray,cv2.COLOR_RGB2GRAY)

# Manual YCrCb
Y = im_np_ycrcb[:, :, 0]
im_np_ycrcb_gray2 = np.array([Y,Y,Y]).transpose(2,1,0)

#print(imnpg)

plt.figure(figsize=(12,8))
plt.subplot(2, 2, 1)
plt.xticks([]),plt.yticks([])
plt.title('RGB')
plt.imshow(im_np_rgb,vmin=0,vmax=255)

plt.subplot(2, 2, 2)
plt.title('RGB -> Grayscale')
plt.imshow(im_np_rgb_gr,cmap='gray',vmin=0,vmax=255)
plt.xticks([]),plt.yticks([])

plt.subplot(2, 2, 3)
plt.title('RGB --> YCrCB --> RGB -> Gray')
plt.imshow(im_np_ycrcb_gray,cmap='gray',vmin=0,vmax=255)
plt.xticks([]),plt.yticks([])

plt.subplot(2, 2, 4)
plt.title('RGB --> YCrCB (Y) --> RGB -> Gray')
plt.imshow(im_np_ycrcb_gray2,cmap='gray',vmin=0,vmax=255)
plt.xticks([]),plt.yticks([])


# Save fig
figimg = plt.gcf() 
figimg.savefig('color_grayscale.png',dpi=300, bbox_inches='tight')

plt.show()

```
