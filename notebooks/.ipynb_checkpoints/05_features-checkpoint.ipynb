# Detect Corners

```python
%load_ext autoreload
%autoreload 2

# put system imports here
import os
import sys
from os.path import join
from glob import glob
from pathlib import Path

# put pip/conda installed imports here
import cv2 as cv
import numpy as np
import pandas as pd
from PIL import Image
import matplotlib.pyplot as plt
import random

# put local imports here
fp_app_root = join(Path(os.getcwd()).parent)
sys.path.append(fp_app_root)
from app.settings import app_cfg
from app.utils import file_utils, draw_utils
from app.models.bbox import BBoxNorm
```

```python
filepaths = glob(join(app_cfg.DIR_IMAGES, '*.jpg'))
fp_im = filepaths[random.randint(0, len(filepaths))]
im_bgr = cv.imread(fp_im)
```

```python
# conver to grayscale
im_gray = cv.cvtColor(im_bgr, cv.COLOR_BGR2GRAY)
```

```python
plt.imshow(im_gray, cmap='gray')
```

```python
corners = cv.goodFeaturesToTrack(im_gray, 20, 0.01, 10)
corners = np.int0(corners)
```

```python
im_dst = im_bgr.copy()

for corner in corners:
  x,y = corner.ravel()
  cv.circle(im_dst, (x,y), 4, (0,255,0), -1, cv.LINE_AA)

plt.figure(figsize=(12,8))
im_dst = cv.cvtColor(im_dst, cv.COLOR_BGR2RGB)
plt.yticks([]),plt.xticks([])
plt.imshow(im_dst),plt.show()
```
