# 02: Image Manipulation

This notebook demonstrates color conversion, resizing

- 02a: RGB, BGR, Grayscale
- 02b: Resize images


```python
import os
import os.path as osp
import cv2
%matplotlib inline
import matplotlib.pyplot as plt
import imutils
```

```python
# Functions are now moved to external file "utils/imx.py"
%reload_ext autoreload
%autoreload 2
from utils import imx # Image function extensions
from config import config as cfg
```

# 02a convert BGR -> RGB

```python
# Read an image and display it
filepath = osp.join(cfg.IMAGES_DIR,'obama.jpg')

# Read and convert to RGB
im = cv2.imread(filepath) #BGR
im = cv2.cvtColor(im,cv2.COLOR_BGR2RGB) # RGB
im = cv2.cvtColor(im,cv2.COLOR_RGB2BGR) # BGR
im = cv2.cvtColor(im,cv2.COLOR_RGB2BGR) # RGB
im = cv2.cvtColor(im,cv2.COLOR_BGR2GRAY) # GRAY
im = cv2.cvtColor(im,cv2.COLOR_GRAY2RGB) # back to RGB
imx.pltimg(im) # should be grayscale with 3 color channels
```

# 02b Resize image

```python
# use the imutils package, which is easier than cv2
im_small = imutils.resize(im,width=60)
imx.pltimg(im_small,mode='bgr')
```

```python
# write image to disk using
fpath_out = osp.join(cfg.OUTPUT_DIR,'ch2_small.jpg')
cv2.imwrite(fpath_out,im_small)
```
