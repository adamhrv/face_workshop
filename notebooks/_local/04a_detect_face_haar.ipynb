# 03: Detecting a Face

Use haarcascades to detect a face in an image


```python
import os
import os.path as osp
import cv2
%matplotlib inline
import matplotlib.pyplot as plt
# Functions moved to external file "utils/imx.py"
%reload_ext autoreload
%autoreload 2
from utils import imx # Image function extensions
from config import config as cfg
```

Haarcascades are an efficient way to detect a face in an image. There are 4 pretrained cascades that are included with OpenCV. 

```python
# Define a function to detect faces using OpenCV's haarcascades
def detect_faces(classifier,src,scale_factor=1.1,overlaps=3,
                min_size=70, max_size=700,
                flags=0):
    
    min_size = (min_size, min_size) # minimum face size
    max_size = (max_size, max_size) # maximum face size
    
    # Convert to grayscale
    src_gray = cv2.cvtColor(src,cv2.COLOR_BGR2GRAY)
    
    # Run detector
    matches = classifier.detectMultiScale(src_gray, 
                                          scale_factor, 
                                          overlaps, 
                                          flags, 
                                          min_size, 
                                          max_size)
    # By default, OpenCV returns x,y,w,w
    # Modify to return x1,y1,x2,y2 (more convenient for drawing)
    matches = [ (r[0],r[1],r[0]+r[2],r[1]+r[3]) for r in matches]
    
    return matches
```

```python
# Load each of the cascades
#cascade = 'haarcascade_frontalface_default.xml'
cascade = 'haarcascade_frontalface_alt2.xml'
classifier = cv2.CascadeClassifier(osp.join(cfg.MODELS_DIR,'opencv',cascade))
```

# 02a Detect Frontal Face

```python
# Read an image and display it
fpath = osp.join(cfg.IMAGES_DIR,'features/IMG_1262_crop.JPG')
im = cv2.imread(fpath)
```

```python
# detect faces
faces = detect_faces(classifier,im)
print('Found {} face(s)'.format(len(faces)))
print("'face' is type: {}".format(type(faces)))
x1,y1,x2,y2=faces[0]
w,h = (x2-x1,y2-y1)
print('First face is: x1: {}, y1: {}, x2: {}, y2: {}'.format(x1,y1,x2,y2))
print("Width: {}, height: {}".format(w,h))
```

```python
# Draw all faces
dst = im.copy() # make copy of original image to draw on

# Iterate through, and draw rectangle for each face
for face in faces:
    x1,y1,x2,y2 = face
    cv2.rectangle(dst,(x1,y1),(x2,y2),(0,255,0),3)

title='HEY! Detect: {} detections'.format(len(faces))
imx.pltimg(dst,mode='bgr',title=title)
```

```python
# Try again, but change detector options
faces = detect_faces(classifier,im,overlaps=0,
                     min_size=40,max_size=200,
                     scale_factor=1.01)

print('Found {} face(s)'.format(len(faces)))

dst = im.copy()
for face in faces:
    x1,y1,x2,y2 = face
    p1 = (x1,y1)
    p2 = (x2,y2)
    cv2.rectangle(dst,p1,p2,(0,255,0),3)
    
title='Haar Detect: 2 overlaps {} detections'.format(len(faces))
imx.pltimg(dst,mode='bgr',title=title,figsize=(12,8))
```

```python
# save the result
fpath_out = osp.join(cfg.OUTPUT_DIR,'ch3_haar_face.png')
cv2.imwrite(fpath_out,dst)
```

```python
# What happens when overlaps is 0?
# 1.5 --> 0.00252700s
# 1.05 --> 0.02325201
import time
st = time.time()
faces = detect_faces(classifier,im,overlaps=5,
                     min_size=100,max_size=500,
                     scale_factor=1.5)
print('Found {} face(s) in {:.8f} seconds'.format(len(faces),time.time()-st))

dst = im.copy()
for face in faces:
    x1,y1,x2,y2 = face
    p1 = (x1,y1)
    p2 = (x2,y2)
    cv2.rectangle(dst,p1,p2,(0,255,0),1)
    
title='Haar Detect: 5 overlaps {} detections'.format(len(faces))
imx.pltimg(dst,mode='bgr',title=title)

# Try other combinations and notice the effect on the speed
# More overlaps means a higher confidence is needed to confirm a face
# A lower scale factor (i.e. 1.05) means the detector will use an aggressive search but is slower
# A higher scale factor (i.e. 1.5) means the detector will use a cursory search and it faster
# Technically, scale factor refers to the multiplying factor for how much the 
# sliding window size increases per scan
# the scan starts at the min size and ends at the max size
# increasing the min size or decreasing the max size will also inrease speed
# conversely, a lower min size or a higher max will will decrease speed
# Exercise: Try "min_size=24" and notice the speed drastically decrease
```
